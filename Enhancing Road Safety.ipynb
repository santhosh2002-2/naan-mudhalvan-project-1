```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd

try:
    df_traffic_accidents = pd.read_csv('traffic_accidents.csv')
    display(df_traffic_accidents.head())
    print(df_traffic_accidents.shape)
except FileNotFoundError:
    print("Error: 'traffic_accidents.csv' not found. Please ensure the file is in the correct location and accessible.")
except pd.errors.EmptyDataError:
    print("Error: 'traffic_accidents.csv' is empty.")
except pd.errors.ParserError:
    print("Error: Unable to parse 'traffic_accidents.csv'. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

import matplotlib.pyplot as plt
import seaborn as sns

# Examine data types
print("Data Types:\n", df_traffic_accidents.dtypes)

# Identify missing values
missing_values = df_traffic_accidents.isnull().sum()
print("\nMissing Values:\n", missing_values)

# Visualize missing values with a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_traffic_accidents.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Descriptive statistics for numerical columns
numerical_cols = df_traffic_accidents.select_dtypes(include=['number'])
print("\nDescriptive Statistics for Numerical Columns:\n", numerical_cols.describe())

# Explore categorical columns
categorical_cols = ['Accident Severity', 'Day of Week', 'Time of Day', 'Road Condition', 'Lighting Conditions']
for col in categorical_cols:
    print(f"\nUnique values and counts for {col}:\n{df_traffic_accidents[col].value_counts()}")
    plt.figure(figsize=(8, 4))  # Adjust figure size as needed
    df_traffic_accidents[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Correlation between numerical features
correlation_matrix = numerical_cols.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Fill missing values in specified columns with 'Unknown'
df_traffic_accidents['Traffic Control Presence'].fillna('Unknown', inplace=True)
df_traffic_accidents['Driver License Status'].fillna('Unknown', inplace=True)

# Handle outliers in numerical columns using IQR method
numerical_cols = ['Number of Vehicles Involved', 'Number of Casualties', 'Driver Age', 'Speed Limit (km/h)']
for col in numerical_cols:
    Q1 = df_traffic_accidents[col].quantile(0.25)
    Q3 = df_traffic_accidents[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_traffic_accidents[col] = df_traffic_accidents[col].clip(lower=lower_bound, upper=upper_bound)

display(df_traffic_accidents.head())

# Correcting chained assignment for filling missing values
df_traffic_accidents['Traffic Control Presence'] = df_traffic_accidents['Traffic Control Presence'].fillna('Unknown')
df_traffic_accidents['Driver License Status'] = df_traffic_accidents['Driver License Status'].fillna('Unknown')

# Convert 'Driver Age' to numeric, coercing errors to NaN
df_traffic_accidents['Driver Age'] = pd.to_numeric(df_traffic_accidents['Driver Age'], errors='coerce')

# Check for inconsistent values in 'Road Condition' and standardize them
df_traffic_accidents['Road Condition'] = df_traffic_accidents['Road Condition'].str.lower()
df_traffic_accidents['Road Condition'] = df_traffic_accidents['Road Condition'].replace({'wet': 'wet', 'dry':'dry'})

display(df_traffic_accidents.head())

import matplotlib.pyplot as plt
import seaborn as sns

# Accident frequency across days of the week
day_of_week_counts = df_traffic_accidents['Day of Week'].value_counts()
plt.figure(figsize=(10, 6))
day_of_week_counts.plot(kind='bar', color='skyblue')
plt.title('Accident Frequency by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

#  Time of day vs. accident severity
plt.figure(figsize=(12, 6))
sns.countplot(x='Time of Day', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Road conditions and accident severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Road Condition', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Driver demographics and accident involvement
plt.figure(figsize=(10, 6))
sns.boxplot(x='Accident Severity', y='Driver Age', data=df_traffic_accidents)
plt.title('Driver Age Distribution by Accident Severity')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Driver Gender', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Driver Gender')
plt.show()

# Speed limits and accident severity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Speed Limit (km/h)', y='Accident Severity', data=df_traffic_accidents, hue='Accident Severity')
plt.title('Speed Limit vs. Accident Severity')
plt.show()

# Summary of key findings
print("Summary of Key Findings:")
print("- Accident frequency by day of week:")
print(day_of_week_counts)
print("\n- The relationship between time of day and accident severity can be visualized in the corresponding plot.")
print("\n- The impact of road conditions on accident severity can be visualized in the corresponding plot.")
print("\n- The influence of driver demographics on accident involvement can be visualized in the corresponding plots.")
print("\n- The correlation between speed limits and accident severity can be visualized in the corresponding plot.")
import matplotlib.pyplot as plt
import seaborn as sns

# Accident Frequency by Day of Week
plt.figure(figsize=(10, 6))
df_traffic_accidents['Day of Week'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Accident Frequency by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Accident Severity by Time of Day
plt.figure(figsize=(12, 6))
sns.countplot(x='Time of Day', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
#Accident Severity by Road Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='Road Condition', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Driver Age Distribution by Accident Severity
plt.figure(figsize=(10, 6))
sns.boxplot(x='Accident Severity', y='Driver Age', data=df_traffic_accidents)
plt.title('Driver Age Distribution by Accident Severity')
plt.show()

# Accident Severity by Driver Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Driver Gender', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Accident Severity by Driver Gender')
plt.show()

# Speed Limit vs. Accident Severity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Speed Limit (km/h)', y='Accident Severity', hue='Accident Severity', data=df_traffic_accidents)
plt.title('Speed Limit vs. Accident Severity')
plt.show()

```
